```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 불러오기
data = pd.read_csv("C:/Users/PC/onedrive(yonsei)/문서/GitHub/YBIGTA_newbie_team_project/database/preprocessed_reviews_naver.csv")

# 날짜별 리뷰 수 확인
data['date'] = pd.to_datetime(data['date'])
daily_reviews = data.groupby(data['date'].dt.date).size()

# 라인 그래프 그리기
plt.figure(figsize=(12, 6))
plt.plot(daily_reviews.index, daily_reviews.values, marker='o', label='# of review')
plt.title("# of review per date")
plt.xlabel("date")
plt.ylabel("# of review")
plt.grid()
plt.legend()
plt.show()
```

```python
from collections import Counter

# 키워드 추출
keywords = ['아이언맨', '마블', '어벤져스', '영화', '재미', '완벽한']
text_data = ' '.join(data['review'].astype(str))  # 리뷰 전체 텍스트 합치기
keyword_counts = Counter([word for word in text_data.split() if word in keywords])

# 막대 그래프
plt.figure(figsize=(10, 5))
plt.bar(keyword_counts.keys(), keyword_counts.values(), color='skyblue')
plt.title("keyword frequency")
plt.xlabel("keyword")
plt.ylabel("frequency")
plt.show()
```

```python
import seaborn as sns

# 산점도: 평점 vs 리뷰 길이
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['rating'], y=data['review_length'], alpha=0.6)
plt.title("평점과 리뷰 길이의 관계")
plt.xlabel("평점")
plt.ylabel("리뷰 길이")
plt.show()
```

```python
from wordcloud import WordCloud

# 워드클라우드 생성
wc = WordCloud(font_path='malgun.ttf', background_color='white', width=800, height=400)
wc.generate(text_data)

# 시각화
plt.figure(figsize=(12, 6))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.title("리뷰 워드클라우드")
plt.show()
```
